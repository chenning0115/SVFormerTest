```python
import os, sys, json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
prefix = "../res_patch_size/"
name = []
keys = []
checks = []
splits = []
oas = []
max_oas = []
for s in os.listdir(prefix):
    # print(s)
    if 'json' in s[-20:]:
        ll = s.split('_')
        # print(ll)
        name.append(ll[0])
        if "tranbase" in s and len(ll)==6:
            # print(s)
            keys.append("".join(ll[2:4]))
            splits.append(ll[4])
        else:
            keys.append(ll[2])
            splits.append(ll[3])
        checks.append(len(ll[2].split('-'))==4)
        jj = json.loads(open("%s/%s" %(prefix, s)).read())
        oa = jj['eval']['oa']
        oas.append(oa)
        if 'max_oa' in jj:
            max_oas.append(jj['max_oa']['value'][-1])
        else:
            max_oas.append(-1)


dd = {
    "data": name,
    "key": keys,
    "check": checks,
    "split": splits,
    "oa": oas,
    "max_oa": max_oas
}
df = pd.DataFrame(dd)
```

```python
# df = df[df['check']]
df = df[['data', 'key', 'split', 'oa', 'max_oa']]
```
