```python
import os, sys, json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def parse_acc(ss):
    ff = []
    for s in ss.split(" "):
        s = s.strip('\n').strip('.').strip('[').strip(']')
        if '' == s:
            continue
        ff.append(float(s))
    return ff
```

```python
def parse_json(jj):
    oa = jj['eval']['oa']
    aa = jj['eval']['aa']
    kappa = jj['eval']['kappa']

    acc = parse_acc(jj['eval']['each_acc'])
    class_num = jj['param']['data']['num_classes']
    assert len(acc) == class_num
    return oa, aa, kappa, acc
```

```python
prefix = "../res_nums/"
data = []
method = []
num = []
split = []
n = []
oas = []
aas = []
kappas = []
vals = {i:[] for i in range(30)}


def add_item(items):
    data.append(items['idata'])
    method.append(items['imethod'])
    num.append(items['inum'])
    split.append(items['isplit'])
    n.append(items['in'])
    oa, aa, kappa, acc = items['oa'], items['aa'], items['kappa'], items['acc']
    oas.append(oa)
    aas.append(aa)
    kappas.append(kappa)
    for i, v in enumerate(acc):
        vals[i].append(v)
    for j in range(len(acc), 30):
        vals[j].append(-1)
    

for s in os.listdir(prefix):
    print(s)
    if 'json' in s[-20:]:
        ll = s.split('_')
        jj = json.loads(open("%s/%s" %(prefix, s)).read())
        oa, aa, kappa, acc = parse_json(jj)
        items = {
            'idata': ll[0],
            'imethod': ll[1],
            'inum': int(ll[2].split('-')[1]),
            'isplit': ll[3],
            'in': int(ll[4][:-4]),
            'oa': oa,
            'aa': aa,
            'kappa': kappa,
            'acc': acc
        }
        add_item(items)


dd = {
    "data_sign": data,
    "method": method,
    "sample_num": num,
    "split": split,
    "oa": oas,
    "aa": aas,
    "kappa": kappas
}
for k, v in vals.items():
    dd[str(k)] = v
df = pd.DataFrame(dd)
```
